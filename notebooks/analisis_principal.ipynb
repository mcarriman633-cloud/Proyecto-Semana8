```python
import pandas as pd
import os
import glob
from IPython.display import display 

#Definir la ruta
folder_path = r"C:\Users\Anty\Documents\GitHub\IACC\Proyecto-Semana8\data"

#Buscar el archivo
files = glob.glob(os.path.join(folder_path, "dataset*.csv"))
if not files:
    files = glob.glob(os.path.join(folder_path, "dataset*.xlsx"))

if files:
    ruta = files[0]
    print(f"Archivo detectado: {os.path.basename(ruta)}")
    
    #Cargar
    if ruta.endswith('.csv'):
        df = pd.read_csv(ruta)
    else:
        df = pd.read_excel(ruta)
    
    #Procesar y mostrar
    df['fecha_registro'] = pd.to_datetime(df['fecha_registro'])
    display(df.head()) 
else:
    print("Error: No se encontró el archivo.")
```
